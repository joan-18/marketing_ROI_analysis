```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Regression
import statsmodels.api as sm

# Statistics
from scipy.stats import shapiro
```

```python
# Load dataset
df = pd.read_csv("marketing.csv")

# Display first five rows
df.head()
```

```python
# Dataset information
df.info()

# Summary statistics
df.describe()

# Check shape
print("Rows and Columns:", df.shape)
```

```python
# Missing values
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="Blues",
            fmt=".2f")

plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(18,5))

sns.scatterplot(x='TV', y='Sales', data=df, ax=axes[0])
axes[0].set_title("TV vs Sales")

sns.scatterplot(x='Radio', y='Sales', data=df, ax=axes[1])
axes[1].set_title("Radio vs Sales")

sns.scatterplot(x='Social_Media', y='Sales', data=df, ax=axes[2])
axes[2].set_title("Social Media vs Sales")

plt.tight_layout()
plt.show()
```

```python
correlations = df.corr(numeric_only=True)['Sales'].sort_values(ascending=False)

print(correlations)
```

```python
X = df[['TV']]
y = df['Sales']
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
predictions = model.predict(X)

predictions.head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df['TV'], y, alpha=0.7)

plt.plot(df['TV'],
         predictions,
         color='red',
         linewidth=3)

plt.xlabel("TV Advertising")
plt.ylabel("Sales")
plt.title("Simple Linear Regression")

plt.show()
```

```python
residuals = y - predictions
```

```python
plt.figure(figsize=(8,6))

plt.scatter(predictions, residuals)

plt.axhline(y=0,
            color='red',
            linestyle='--')

plt.xlabel("Predicted Sales")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.histplot(residuals,
             kde=True)

plt.title("Residual Distribution")

plt.show()
```

```python
sm.qqplot(residuals,
          line='45')

plt.title("QQ Plot")

plt.show()
```

```python
stat, p = shapiro(residuals)

print("Statistic:", stat)
print("P-value:", p)
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(x=predictions,
                y=residuals)

plt.axhline(0,
            color='red',
            linestyle='--')

plt.xlabel("Predicted Values")
plt.ylabel("Residuals")

plt.title("Homoscedasticity Check")

plt.show()
```

ROI Recommendation


The analysis shows that TV advertising has the strongest positive relationship with sales. The regression model explains a large proportion of sales variability (high R²), and the TV coefficient is statistically significant (p < 0.05). Therefore, increasing investment in TV advertising is expected to generate the greatest return on marketing spend compared with the other channels analyzed.
